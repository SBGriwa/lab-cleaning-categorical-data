```python
import pandas as pd
import numpy as np
import datetime
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

# Load the csv. Use the variable customer_df as customer_df = pd.read_csv()

```python
customer_df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
customer_df
```

```python
customer_df.columns = [customer_df.columns[i].lower().replace(' ', '_') for i in range(len(customer_df.columns)) ]
 
customer_df.columns
```

# What should we do with the customer_id column?

```python
customer_df = customer_df.drop(['customer'], axis=1)
customer_df
```

# Load the continuous and discrete variables into numericals_df and categorical_df variables

```python
categorical_df = customer_df.select_dtypes('object')
categorical_df
```

```python
customer_df['effective_to_date']
```

```python
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'], errors='coerce')
customer_df['effective_to_date']
```

```python
numerical_df = customer_df.select_dtypes('number')
numerical_df
```

# Plot every categorical variable. What can you see in the plots? 

```python
for column in categorical_df.columns:
    plt.figure(figsize=(12,6))
    sns.barplot(x=column, y="total_claim_amount", data=customer_df)
    plt.show()
```

# For the categorical data, check if there is any data cleaning that need to perform. 

```python
for column in categorical_df.columns:
    print(customer_df[column].value_counts())
```

```python
customer_df['employmentstatus'].value_counts()
```

```python
customer_df['employmentstatus'] = customer_df['employmentstatus'].apply(lambda x: x if x in ['Employed', 'Unemployed'] else 'Other')

customer_df['employmentstatus'].value_counts()
```

```python
customer_df['policy'].value_counts()
```

```python
customer_df['policy'] = customer_df['policy'].apply(lambda x: 'Special' if x in ['Special L3','Special L2','Special L1'] else x)

customer_df['policy'].value_counts()
```

```python
customer_df['vehicle_class'].value_counts()
```

```python
customer_df['vehicle_class'] = customer_df['vehicle_class'].apply(lambda x: 'Luxury' if x in ['Luxury SUV', 'Luxury Car'] else x)

customer_df['vehicle_class'].value_counts()
```
